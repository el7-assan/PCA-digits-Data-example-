```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_digits
dataset = load_digits()
dataset
```

```python
dataset.keys()
```

```python
dataset.data.shape
```

```python
np.unique(dataset.target)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(6, 6))  # حجم صغير ومناسب
for i, ax in enumerate(axes.flat):
    ax.imshow(dataset.data[i].reshape(8, 8), cmap='gray')
    ax.set_title(f"Image {i}")
    
plt.tight_layout()
plt.show()
```

```python
data=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
data.head()
```

```python
data.describe()
```

```python
x=data
y= dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

# PCA algorithm

```python
x.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)

x_pca = pca.fit_transform(x)

x_pca.shape
```

 from 64 columns to 29 

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_ # how many components 
```

```python
# second training
x_train_pca,x_test_pca,y_train,y_test = train_test_split(x_pca,y,test_size=0.2,random_state=30)
model =LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train)
model.score(x_test_pca,y_test)

```

Almost the same accuracy, although we reduced the number of columns.
